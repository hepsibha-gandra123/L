```python
movies['label']=movies['Rating']>=7
movies['label']=movies['label'].astype(int)
```

```python
movies.head()
```

```python
movies['label']=movies['label'].map({
    1:'High Rated',
    0:'Low Rated'
})
```

```python
movies.head()
```

```python
movies=movies.drop(columns=['Title','Description'])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
movies['label']=le.fit_transform(movies['label'])
```
